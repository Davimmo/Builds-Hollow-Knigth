```python
# Função que retorna todos os subconjuntos que somam a um dado valor
def findSubsets(books, n, target_sum, current_subset, all_subsets):
    # Caso base: se o target_sum for 0, adiciona a combinação atual
    if target_sum == 0:
        all_subsets.append(current_subset[:])  # Adiciona uma cópia do subconjunto atual
        return

    # Se não houver mais livros ou o target_sum for negativo, retorne
    if n == 0 or target_sum < 0:
        return

    # Ignorar o último livro e continuar com os n-1 livros restantes
    findSubsets(books, n - 1, target_sum, current_subset, all_subsets)

    # Incluir o último livro e continuar com os n-1 livros restantes
    current_subset.append(books[n - 1])  # Adiciona o último livro ao subconjunto atual
    findSubsets(books, n - 1, target_sum - books[n - 1][1], current_subset, all_subsets)

    # Após a recursão, remover o último livro para voltar à etapa anterior
    current_subset.pop()

# Função principal para encontrar todas as combinações de livros que somam o valor desejado
def findBooksWithSum(books, target_sum):
    all_subsets = []  # Lista para armazenar todos os subconjuntos válidos
    findSubsets(books, len(books), target_sum, [], all_subsets)  # Chama a função recursiva
    return all_subsets

# Exemplo de livros com pesos (nome, peso)
books = [
    ("Bússola Caprichosa", 1),
    ("Enxame de Colecionadores", 1),
    ("Carapaça Robusta", 2),
    ("Apanhador de Almas", 2),
    ("Pedra do Xamã", 3),
    ("Devorador de Almas", 4),
    ("Mestre da Esquiva", 2),
    ("Mestre da Corrida", 1),
    ("Canção das Larvas", 1),
    ("Elegia da Larvamosca", 3),
    ("Coração Frágil", 2),
    ("Coração Inquebrável", 2),
    ("Ganância Frágil", 2),
    ("Ganância Inquebrável", 2),
    ("Força Frágil", 3),
    ("Força Inquebrável", 3),
    ("Dobrador de Magias", 2),
    ("Corpo Firme", 1),
    ("Golpe Pesado", 2),
    ("Corte Rápido", 3),
    ("Ferrão Longo", 2),
    ("Marca de Orgulho", 3),
    ("Fúria dos Caídos", 2),
    ("Espinhos da Agonia", 1),
    ("Carapaça de Baldur", 2),
    ("Ninho de Flukes", 3),
    ("Insignia do Defensor", 1),
    ("Útero Brilhante", 2),
    ("Foco Rápido", 3),
    ("Foco Profundo", 4),
    ("Coração de Sangue Vital", 2),
    ("Núcleo de Sangue Vital", 3),
    ("Bênção de Joni", 4),
    ("Sangue da Colmeia", 4),
    ("Cogumelo com Esporos", 1),
    ("Sombra Afiada", 2),
    ("Forma de Unn", 2),
    ("Glória do Mestre do Ferrão", 1),
    ("Canção das Tecelãs", 2),
    ("Portador dos Sonhos", 1),
    ("Escudo dos Sonhos", 3),
    ("Criança Grimm", 2),
    ("Melodia Despreocupada", 3),
    ("Alma do Rei", 5)
]

# Valor desejado
target_sum = 11

# Encontra todas as combinações de livros que somam o valor desejado
valid_combinations = findBooksWithSum(books, target_sum)

# Exibe o resultado
print(len(valid_combinations))
print(valid_combinations[1])
```

```python
from collections import defaultdict

def verificar_primeiras_duplicatas(lista_de_listas, max_exemplos=5):
    # Dicionário para contar as listas
    contagem = defaultdict(int)
    exemplos_mostrados = 0  # Contador para o número de duplicatas exibidas
    
    # Normalizar (ordenar) e contar cada lista
    for lista in lista_de_listas:
        # Ordena a lista para ignorar a ordem dos elementos
        lista_ordenada = tuple(sorted(lista))  # Convertemos em tupla, pois listas não são "hashable"
        contagem[lista_ordenada] += 1
        
        # Exibir a duplicata se aparecer mais de uma vez
        if contagem[lista_ordenada] == 2:  # Conta só quando encontra a segunda ocorrência (primeira duplicata)
            print(f"Lista {lista_ordenada} apareceu pela segunda vez.")
            exemplos_mostrados += 1
        
        # Para após exibir as primeiras 5 duplicatas
        if exemplos_mostrados >= max_exemplos:
            break

    if exemplos_mostrados == 0:
        print("Nenhuma lista duplicada foi encontrada.")

# Exemplo de lista de listas
lista_de_listas = combinacoes_validas

verificar_primeiras_duplicatas(lista_de_listas, max_exemplos=5)
```
